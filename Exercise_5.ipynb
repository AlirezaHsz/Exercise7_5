        <li>
        ستون "Dt_Customer" شامل تاریخ عضویت مشتریان است. ابتدا این ستون را به فرمت تاریخ (datetime) تبدیل می‌کنیم. سپس فقط بخش تاریخ (بدون زمان) را استخراج می‌کنیم. در نهایت، جدیدترین و قدیمی‌ترین تاریخ ثبت‌نام مشتریان را برای بررسی اولیه نمایش می‌دهیم.

```python
# TODO: تبدیل به دیتاتایپ Dt_Customer و استخراج ویژگی

# تبدیل ستون تاریخ ثبت‌نام به فرمت datetime
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], format="%d-%m-%Y")

# نمایش جدیدترین و قدیمی‌ترین تاریخ ثبت‌نام مشتریان
print("جدیدترین تاریخ ثبت‌نام:", df["Dt_Customer"].max())
print("قدیمی‌ترین تاریخ ثبت‌نام:", df["Dt_Customer"].min())

```

<br>
<div dir="rtl">
    <font size="4">
        <li>
            یک ویژگی جدید به نام Customer_for  ایجاد کنید.
         ویژگی <strong>Customer_For</strong> نشان‌دهنده مدت‌زمانی است که هر مشتری از فروشگاه خرید می‌کند، بر اساس اختلاف تاریخ عضویت او با جدیدترین تاریخ ثبت‌شده در داده‌ها. </li> <li> برای این کار، ابتدا تاریخ عضویت مشتریان را به فرمت <strong>تاریخی</strong> تبدیل کنید. سپس جدیدترین تاریخ موجود در داده‌ها را مشخص کرده و اختلاف این دو تاریخ را برحسب روز محاسبه کنید.
        </li>
</font>
</div>

```python
# TODO: ایجاد ویژگی Customer_For

# محاسبه مدت‌زمان عضویت مشتری از تاریخ ثبت‌نام تا جدیدترین تاریخ موجود
latest_date = df["Dt_Customer"].max()
df["Customer_For"] = (latest_date - df["Dt_Customer"]).dt.days
print(df[["Dt_Customer", "Customer_For"]].head())
```

    <font size="4">
        <li>
                داده های پرت (Outlier) هارا به روش IQR حذف کنید
        </li>
        <li>
            <font color="red">نکته مهم:</font> استفاده از مقادیر پیشفرض Q1 و Q3 داده پیشنهاد نمی شود. این دیتاست داده های کمی دارد و حذف تعداد زیادی از داده ها توصیه نمی شود. شما می توانید از مقادیر کمتری/بیشتری برای Q1 و Q3 استفاده کنید. مثلا مقدار 0.20 به جای 0.25 برای Q1 و مقدار 0.80 برای Q3.(مقادیر را خودتان انتخاب کنید)
        </li>
    </font>
</div>
<br>

```python
# TODO: حذف داده های پرت
def remove_outliers_iqr(df, cols, q1=0.20, q3=0.80):
    """
    حذف داده‌های پرت بر اساس روش IQR با مقادیر قابل تنظیم برای Q1 و Q3.
    """
    for col in cols:
        Q1 = df[col].quantile(q1)  # محاسبه Q1 با مقدار تنظیم‌شده
        Q3 = df[col].quantile(q3)  # محاسبه Q3 با مقدار تنظیم‌شده
        IQR = Q3 - Q1  # محاسبه دامنه بین‌چارکی
        lower_bound = Q1 - 1.5 * IQR  # مرز پایین
        upper_bound = Q3 + 1.5 * IQR  # مرز بالا

        # حذف داده‌های پرت
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

    return df

# لیست ستون‌هایی که باید داده‌های پرت آن‌ها حذف شود
cols_to_clean = ["Income", "Wines", "Fruits", "Meat", "Fish", "Sweets", "Gold", "Total_Purchases", "Spent"]

# اعمال تابع حذف داده‌های پرت
df_cleaned = remove_outliers_iqr(df, cols_to_clean)

# نمایش اطلاعات پس از حذف داده‌های پرت
print(df_cleaned.shape)
```

```python
# TODO: Histogram رسم
import matplotlib.pyplot as plt
import seaborn as sns

def plot_numeric_histograms(df):
    """
    رسم هیستوگرام برای همه ویژگی‌های عددی دیتافریم

    :param df: دیتافریم مورد بررسی
    """
    numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

    plt.figure(figsize=(12, 8))

    for i, col in enumerate(numeric_cols):
        plt.subplot(len(numeric_cols) // 3 + 1, 3, i + 1)
        sns.histplot(df[col], bins=30, kde=True)
        plt.title(f"توزیع {col}")

    plt.tight_layout()
    plt.show()

# فراخوانی تابع برای رسم هیستوگرام
plot_numeric_histograms(df)
```

```python
# TODO: مقیاس بندی ویژگی ها

# انتخاب فقط ویژگی‌های عددی برای Scaling
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

# انجام مقیاس‌بندی
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[numeric_cols]), columns=numeric_cols)

# نمایش چند مقدار اول پس از مقیاس‌بندی
df_scaled.head()
```

```python
# محاسبه همبستگی بین همه ویژگی‌ها
corr_matrix = df_scaled.corr()

# انتخاب آستانه‌ی حداقلی برای حذف ویژگی‌ها (مثلاً 0.05)
threshold = 0.05

# پیدا کردن ویژگی‌هایی با همبستگی ضعیف با Total_Purchases و Spent
low_correlation_features = [col for col in df_scaled.columns
                            if abs(corr_matrix["Total_Purchases"][col]) < threshold and
                               abs(corr_matrix["Spent"][col]) < threshold]

print("ویژگی‌هایی با همبستگی کمتر از 0.05 که می‌توان حذف کرد:", low_correlation_features)

# حذف ویژگی‌های کم‌اهمیت
df_scaled = df_scaled.drop(columns=low_correlation_features)

# نمایش ستون‌های باقی‌مانده پس از حذف
print("ویژگی‌های باقی‌مانده:", df_scaled.columns)
```

```python
# TODO: KMeans  پیدا کردن تعداد کلاستر های مناسب

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# تست تعداد خوشه‌های مختلف و محاسبه مجموع مربعات خطا (inertia)
inertia = []
K_range = range(2, 11)  # تست تعداد خوشه‌های بین 2 تا 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# رسم نمودار روش Elbow برای تعیین تعداد بهینه خوشه‌ها
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.xlabel("تعداد خوشه‌ها")
plt.ylabel("مجموع مربعات خطا (Inertia)")
plt.title("روش Elbow برای تعیین تعداد خوشه‌های مناسب")
plt.show()
```

```python
# TODO :  KMeans پیاده سازی

# انتخاب تعداد بهینه خوشه‌ها
optimal_k = 6  # تعداد بهینه‌ای که از نمودار Elbow انتخاب شده

# اجرای KMeans
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_scaled["Cluster"] = kmeans.fit_predict(df_scaled)

# نمایش تعداد اعضای هر خوشه
print(df_scaled["Cluster"].value_counts())
```

```python
# TODO: GMM پیدا کردن تعداد کلاستر های مناسب
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# تست تعداد خوشه‌های مختلف و محاسبه BIC و AIC
K_range = range(2, 11)
bics, aics = [], []

for k in K_range:
    gmm = GaussianMixture(n_components=k, random_state=42)
    gmm.fit(df_scaled)
    bics.append(gmm.bic(df_scaled))
    aics.append(gmm.aic(df_scaled))

# رسم نمودار BIC و AIC برای تعیین تعداد خوشه‌های مناسب
plt.figure(figsize=(10, 5))
plt.plot(K_range, bics, marker='o', linestyle='--', label="BIC")
plt.plot(K_range, aics, marker='s', linestyle='--', label="AIC")
plt.xlabel("تعداد خوشه‌ها")
plt.ylabel("مقدار BIC / AIC")
plt.title("انتخاب تعداد خوشه‌های بهینه برای GMM")
plt.legend()
plt.show()

```

```python
#TODO: GMM پیاده سازی

optimal_k = 6  # عدد بهینه بر اساس تحلیل BIC/AIC

# اجرای GMM
gmm = GaussianMixture(n_components=optimal_k, random_state=42)
df_scaled["Cluster_GMM"] = gmm.fit_predict(df_scaled)

# نمایش تعداد اعضای هر خوشه
print(df_scaled["Cluster_GMM"].value_counts())
```

## ارزیابی و نام گذاری کلاستر ها (Labeling):
*ارزیابی خودتان از کد و خروجی حاصل از آن را بنویسید*


خوشه 0: مشتریانی با درآمد پایین و هزینه کم → احتمالا مشتریان غیرفعال یا با خرید محدود

خوشه 1: مشتریانی با درآمد متوسط و هزینه متوسط به بالا → مشتریان معمولی که خرید منطقی دارند

خوشه 2: مشتریانی با درآمد بالا و هزینه زیاد → مشتریان VIP با خریدهای سنگین

خوشه 3: مشتریانی با درآمد کم اما خرید کمی بیشتر از حد متوسط → ممکن است مشتریان وفادار با بودجه محدود باشند

خوشه 4: مشتریانی با درآمد و هزینه نزدیک به متوسط → مشتریانی که رفتار خرید متعادلی دارند

خوشه 5: مشتریانی با درآمد پایین و کمترین هزینه خرید → احتمالاً کم‌فعال‌ترین گروه مشتریان

 چگونه این خوشه‌ها را نام‌گذاری کنیم؟

خوشه 0: مشتریان کم‌فعال

 خوشه 1: مشتریان معمولی

 VIP خوشه 2: مشتریان

خوشه 3: مشتریان با بودجه محدود

خوشه 4: مشتریان متعادل

 خوشه 5: مشتریان غیرمشتری یا بسیار کم‌فعال
